```python
from sga import *
```

### Running SGA with a sine fitness function

```python
generations=simple_genetic_algorithm(pop_size=100,
        max_gen=10,
        cross_over_rate=0.9,                        
        mutation_rate=0.05,  
        genome_length=101, 
        fitness_function=fitness_function_sine,
        parent_selection_function=fittest_parent_selection)
```

```python
#plotting the results

x= np.arange(0, 128, 0.001)
y= np.sin(x)
    
scaled_genomes=[]
fitness_values=[]


for individual in generations[0]:
        
        fitness_values.append(individual.fitness_val)
        

        factor=2**(len(individual.genome)-7)
        scaled_genomes.append(int(individual.genome, 2)/factor)
        
plt.plot(x, y)
plt.scatter(scaled_genomes, fitness_values, color='black')
#plt.savefig('generation '+str(10)+".png")
```
